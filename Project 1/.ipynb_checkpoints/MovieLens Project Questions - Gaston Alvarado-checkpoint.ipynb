### 9. Find the top 25 movies according to average ratings such that each movie has number of ratings more than 100 - 10 marks

Hint : 
1. First find the movies that have more than 100 ratings(use merge, groupby and count). Extract the movie titles in a list.
2. Find the average rating of all the movies and sort them in the descending order. You will have to use the .merge() function to reach to a data set through which you can get the names and the average rating.
3. Use isin(list obtained from 1) to filter out the movies which have more than 100 ratings.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
#======================================================================================
#             Merge the two dataframes to integrate ratings and movies
#======================================================================================

#Check the number of rows of each dataframe before and after the merge
print(f'Rows in data dataframe',len(data))
print(f'Rows in item dataframe',len(item))

#Merge both dataframes
df_merge_col = pd.merge(data, item, on='movie id')

print(f'Rows in merged dataframe',len(df_merge_col))

#======================================================================================
#             Create a list for movies with more than 100 reviews
#======================================================================================

#Group by the ratings column
group_by_ratings = df_merge_col.groupby('movie title').count()

#Filter by movies that have more than 100 ratings
more_than_100_ratings = group_by_ratings[group_by_ratings['rating'] > 100]

#Save into a list movies with more than 100 ratings
movies_with_more_than_100_ratings = more_than_100_ratings.index

#======================================================================================
#             Create a list with avg ratings for all movies
#======================================================================================

#Group the list and sort descending
avg_ratings = df_merge_col.groupby('movie title').mean().sort_values(by='rating', ascending=False)

#Reset the index
avg_ratings.reset_index(inplace=True)

#======================================================================================
#             Match the movies that are in the 100+ ratings list
#======================================================================================

#Filter the list using the 100+ ratings list created before
filtered_movies_with_ratings = avg_ratings['movie title'][avg_ratings['movie title'].isin(movies_with_more_than_100_ratings)].to_frame()
filtered_movies_with_ratings

#======================================================================================
#             Match the movies that are in the 100+ ratings list
#======================================================================================

#Merge the two frames and filter the list of the top 25 movies
x = pd.merge(filtered_movies_with_ratings, avg_ratings, on='movie title')
top_25_movies = x[['movie title', 'rating']].head(25)
top_25_movies
```

The data provided is not well balanced since 75% are men, and the catefories of movies are also unbalanced, so we can't really say the numbers and assumptions are accurate, but keeping that in mind, and just by a simple percentage calculation, we can say that our statements are valid, but with a low level of confidence.

```python
#Merge all dataframes
merge_data_user = pd.merge(df_merge_col, user, on='user id')
merge_data_user
```

```python
#Create both groupby objects to collect totals by gender and ratings
g = merge_data_user.groupby('gender').sum()
r = merge_data_user.groupby('gender').count()
r
```

##### Analysis preferences Drama

```python
male_total_ratings = r['rating'][1]
print(f'Total Male that provided ratings', male_total_ratings)

female_total_ratings = r['rating'][0]
print(f'Total Female that provided ratings', male_total_ratings)

male_watch_drama = g['Drama'][1]
print(f'Total Male that watch drama', male_watch_drama)
female_watch_drama = g['Drama'][0]
print(f'Total Female that watch drama', female_watch_drama)

percentage_male_watch_drama =  male_watch_drama/male_total_ratings
print(f'Percentage of Male that watch drama',percentage_male_watch_drama)

percentage_female_watch_drama =  female_watch_drama/female_total_ratings
print(f'Percentage of Female that watch drama',percentage_female_watch_drama)
```
